```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('log_transformed.csv')
```

```python
data.head()
```

```python
y = data.SalePrice.copy()
x = data.drop(['Id','SalePrice'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu', input_dim = 82 ))
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu' ))
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu' ))
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu' ))
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu' ))
model.add(Dense(units=20, kernel_initializer = 'he_uniform', activation = 'relu' ))
model.add(Dense(units=1, kernel_initializer = 'glorot_uniform', activation = 'linear' ))
```

```python
model.summary()
```

```python
import os
import keras
```

```python
callbacks = keras.callbacks.ModelCheckpoint(filepath=os.getcwd(), monitor='val_loss', save_best_only=True)
```

```python
model.compile(optimizer='adam', loss = 'MeanSquaredError', metrics=['mean_squared_error'])
```

```python
model_history = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size = 10, epochs = 250, callbacks=callbacks)
```

```python
model.load_weights(os.getcwd())
```

```python
mean_squared_error(y_train, model.predict(x_train))
```
